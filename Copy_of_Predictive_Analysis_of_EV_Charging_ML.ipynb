```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/content/drive/MyDrive/station_data_dataverse.csv'

ev_df = pd.read_csv(file_path)
ev_df
```

```python
print(ev_df.info())
```

```python
print(ev_df.describe())
```

```python
ev_df.isnull().sum()
```

```python
# Distribution of 'chargeTimeHrs' (target variable)
plt.figure(figsize=(10, 6))
sns.histplot(ev_df['chargeTimeHrs'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Charge Time (Hrs)')
plt.xlabel('Charge Time (Hours)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(y=ev_df['chargeTimeHrs'], color='lightgreen')
plt.title('Box Plot of Charge Time (Hours)')
plt.ylabel('Charge Time (Hours)')
plt.show()
```
